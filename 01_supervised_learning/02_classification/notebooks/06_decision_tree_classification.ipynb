## üìä Dataset Loading & Tree Learning Assessment

**Loading our dataset with a focus on tree-based learning suitability and feature analysis**

````python
# Decision Tree Optimized Dataset Loading
print("üìä DECISION TREE DATASET LOADING & ANALYSIS")
print("="*50)

# Load dataset with tree learning focus
print("üå≥ Loading Social Network Ads dataset for tree analysis...")

try:
    dataset = pd.read_csv('../01_Data/Social_Network_Ads.csv')
    print("‚úÖ Dataset loaded successfully from ../01_Data/")
except FileNotFoundError:
    print("‚ö†Ô∏è  Trying alternative path...")
    try:
        dataset = pd.read_csv('Social_Network_Ads.csv')
        print("‚úÖ Dataset loaded from current directory")
    except FileNotFoundError:
        print("üìù Creating demonstration dataset optimized for Decision Trees...")
        # Create synthetic data with clear hierarchical patterns
        np.random.seed(42)
        n_samples = 400
        
        # Generate data with tree-friendly patterns (clear splits)
        # Create age groups with distinct behaviors
        young_adults = np.random.normal(25, 3, n_samples//4)  # 18-32
        adults = np.random.normal(35, 4, n_samples//4)        # 28-42
        middle_aged = np.random.normal(45, 5, n_samples//4)   # 35-55
        seniors = np.random.normal(55, 4, n_samples//4)       # 48-62
        
        ages = np.concatenate([young_adults, adults, middle_aged, seniors])
        ages = np.clip(ages, 18, 60).astype(int)
        
        # Create salary groups with clear patterns
        salaries = np.zeros(n_samples)
        for i, age in enumerate(ages):
            if age < 30:
                salaries[i] = np.random.normal(35000, 8000)
            elif age < 40:
                salaries[i] = np.random.normal(55000, 12000)
            elif age < 50:
                salaries[i] = np.random.normal(75000, 15000)
            else:
                salaries[i] = np.random.normal(85000, 18000)
        
        salaries = np.clip(salaries, 15000, 150000).astype(int)
        
        # Create hierarchical purchase patterns (perfect for trees)
        purchased = np.zeros(n_samples, dtype=int)
        for i in range(n_samples):
            # Tree-like decision rules
            if ages[i] >= 35:  # First split: Age
                if salaries[i] >= 60000:  # Second split: Salary
                    purchased[i] = np.random.choice([0, 1], p=[0.2, 0.8])  # 80% purchase
                else:
                    purchased[i] = np.random.choice([0, 1], p=[0.6, 0.4])  # 40% purchase
            else:  # Young customers
                if salaries[i] >= 50000:  # Different threshold for young
                    purchased[i] = np.random.choice([0, 1], p=[0.4, 0.6])  # 60% purchase
                else:
                    purchased[i] = np.random.choice([0, 1], p=[0.8, 0.2])  # 20% purchase
        
        dataset = pd.DataFrame({
            'User ID': range(1, n_samples + 1),
            'Gender': np.random.choice(['Male', 'Female'], n_samples),
            'Age': ages,
            'EstimatedSalary': salaries,
            'Purchased': purchased
        })
        print("‚úÖ Synthetic hierarchical dataset created for Decision Trees")

# Display basic dataset information
print(f"\nüå≥ DECISION TREE DATASET OVERVIEW:")
print(f"   üìè Dataset shape: {dataset.shape}")
print(f"   üìä Features: {list(dataset.columns[:-1])}")
print(f"   üéØ Target: {dataset.columns[-1]}")

# Show first few rows
print(f"\nüìã SAMPLE DATA:")
print(dataset.head())

# Extract features and target for tree analysis
X = dataset.iloc[:, [2, 3]].values  # Age and EstimatedSalary
y = dataset.iloc[:, -1].values      # Purchased

print(f"\nüéØ TREE LEARNING ASSESSMENT:")
print(f"   üìä Feature matrix shape: {X.shape}")
print(f"   üéØ Target vector shape: {y.shape}")
print(f"   üìà Class balance: {dict(zip(*np.unique(y, return_counts=True)))}")

# Calculate class distribution
class_counts = np.bincount(y)
class_balance = class_counts / len(y)
print(f"   ‚öñÔ∏è  Class balance ratio: {class_balance[0]:.3f} (No) : {class_balance[1]:.3f} (Yes)")

# Tree suitability analysis
print(f"\nüå≥ DECISION TREE SUITABILITY ANALYSIS:")

# 1. Feature types
print(f"   üìä Feature types:")
print(f"      ‚Ä¢ Age: Continuous (excellent for splits)")
print(f"      ‚Ä¢ Salary: Continuous (excellent for splits)")

# 2. Missing values check
missing_values = dataset.isnull().sum()
print(f"   ‚ùì Missing values: {missing_values.sum()} total")
if missing_values.sum() == 0:
    print(f"      ‚úÖ No missing values - perfect for trees!")

# 3. Feature ranges and distributions
print(f"   üìè Feature ranges:")
print(f"      ‚Ä¢ Age: [{X[:, 0].min():.0f}, {X[:, 0].max():.0f}] years")
print(f"      ‚Ä¢ Salary: [${X[:, 1].min():.0f}, ${X[:, 1].max():.0f}]")

# 4. Class separability preview
age_by_class = [X[y==0, 0], X[y==1, 0]]
salary_by_class = [X[y==0, 1], X[y==1, 1]]

print(f"   üîç Feature separability preview:")
print(f"      ‚Ä¢ Age difference: {abs(np.mean(age_by_class[1]) - np.mean(age_by_class[0])):.1f} years")
print(f"      ‚Ä¢ Salary difference: ${abs(np.mean(salary_by_class[1]) - np.mean(salary_by_class[0])):.0f}")

# 5. Optimal split potential
def calculate_potential_splits(feature, target):
    """Calculate number of potential useful splits for a feature"""
    unique_values = np.unique(feature)
    return len(unique_values) - 1

age_splits = calculate_potential_splits(X[:, 0], y)
salary_splits = calculate_potential_splits(X[:, 1], y)

print(f"   üå≤ Split potential:")
print(f"      ‚Ä¢ Age: {age_splits} possible splits")
print(f"      ‚Ä¢ Salary: {salary_splits} possible splits")

print(f"\n‚úÖ TREE READINESS ASSESSMENT:")
print(f"   üå≥ Continuous features: Perfect for threshold splits")
print(f"   üìä Balanced classes: Good for unbiased learning")
print(f"   üéØ No missing values: No preprocessing needed")
print(f"   üìà Clear separability: Trees should perform well")

print(f"\nüéØ Ready for information theory analysis!")
````

````python
# Tree-Optimized Exploratory Data Analysis
print("üîç TREE-OPTIMIZED EXPLORATORY DATA ANALYSIS")
print("="*50)

# Analyze data specifically for Decision Tree insights
print("üå≥ DECISION TREE PERSPECTIVE ANALYSIS")
print("-"*40)

# Class distribution analysis
unique_classes, class_counts = np.unique(y, return_counts=True)
print(f"üìä CLASS DISTRIBUTION:")
for cls, count in zip(unique_classes, class_counts):
    percentage = (count / len(y)) * 100
    print(f"   Class {cls}: {count} samples ({percentage:.1f}%)")

# Feature statistics by class
print(f"\nüìà FEATURE STATISTICS BY CLASS:")
for class_idx in unique_classes:
    mask = y == class_idx
    class_label = "No Purchase" if class_idx == 0 else "Purchased"
    
    print(f"   {class_label} (Class {class_idx}):")
    print(f"      Age: Œº={X[mask, 0].mean():.1f}, œÉ={X[mask, 0].std():.1f}, range=[{X[mask, 0].min():.0f}, {X[mask, 0].max():.0f}]")
    print(f"      Salary: Œº=${X[mask, 1].mean():.0f}, œÉ=${X[mask, 1].std():.0f}, range=[${X[mask, 1].min():.0f}, ${X[mask, 1].max():.0f}]")

# Create comprehensive visualizations for tree analysis
plt.figure(figsize=(20, 15))

# 1. Feature distributions by class (histogram)
plt.subplot(3, 4, 1)
ages_no = X[y==0, 0]
ages_yes = X[y==1, 0]
plt.hist(ages_no, bins=15, alpha=0.7, label='No Purchase', color='red', density=True)
plt.hist(ages_yes, bins=15, alpha=0.7, label='Purchased', color='green', density=True)
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Age Distribution by Class\n(Identify Split Points)')
plt.legend()
plt.grid(True, alpha=0.3)

# Add potential split lines
age_split_candidates = [30, 35, 40, 45]
for split in age_split_candidates:
    if X[:, 0].min() <= split <= X[:, 0].max():
        plt.axvline(split, color='blue', linestyle='--', alpha=0.5)

plt.subplot(3, 4, 2)
salaries_no = X[y==0, 1]
salaries_yes = X[y==1, 1]
plt.hist(salaries_no, bins=15, alpha=0.7, label='No Purchase', color='red', density=True)
plt.hist(salaries_yes, bins=15, alpha=0.7, label='Purchased', color='green', density=True)
plt.xlabel('Estimated Salary')
plt.ylabel('Density')
plt.title('Salary Distribution by Class\n(Identify Split Points)')
plt.legend()
plt.grid(True, alpha=0.3)

# Add potential split lines
salary_split_candidates = [40000, 50000, 60000, 70000, 80000]
for split in salary_split_candidates:
    if X[:, 1].min() <= split <= X[:, 1].max():
        plt.axvline(split, color='blue', linestyle='--', alpha=0.5)

# 2. Box plots for outlier detection
plt.subplot(3, 4, 3)
box_data_age = [ages_no, ages_yes]
box_plot = plt.boxplot(box_data_age, labels=['No Purchase', 'Purchased'], patch_artist=True)
box_plot['boxes'][0].set_facecolor('red')
box_plot['boxes'][1].set_facecolor('green')
plt.ylabel('Age')
plt.title('Age Distribution\n(Outlier Detection)')
plt.grid(True, alpha=0.3)

plt.subplot(3, 4, 4)
box_data_salary = [salaries_no, salaries_yes]
box_plot = plt.boxplot(box_data_salary, labels=['No Purchase', 'Purchased'], patch_artist=True)
box_plot['boxes'][0].set_facecolor('red')
box_plot['boxes'][1].set_facecolor('green')
plt.ylabel('Estimated Salary')
plt.title('Salary Distribution\n(Outlier Detection)')
plt.grid(True, alpha=0.3)

# 3. 2D Feature space (the most important for trees)
plt.subplot(3, 4, 5)
scatter_no = plt.scatter(X[y==0, 0], X[y==0, 1], c='red', alpha=0.6, s=50, label='No Purchase')
scatter_yes = plt.scatter(X[y==1, 0], X[y==1, 1], c='green', alpha=0.6, s=50, label='Purchased')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('2D Feature Space\n(Decision Boundary Preview)')
plt.legend()
plt.grid(True, alpha=0.3)

# Add potential decision boundaries (rectangular splits)
plt.axvline(35, color='blue', linestyle='--', alpha=0.7, linewidth=2, label='Age Split')
plt.axhline(60000, color='orange', linestyle='--', alpha=0.7, linewidth=2, label='Salary Split')

# 4. Correlation analysis
plt.subplot(3, 4, 6)
correlation_matrix = np.corrcoef(X.T)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            xticklabels=['Age', 'Salary'], yticklabels=['Age', 'Salary'])
plt.title('Feature Correlation\n(Independence Check)')

# 5. Feature importance preview (manual calculation)
plt.subplot(3, 4, 7)

# Calculate simple feature importance based on class separation
def calculate_class_separation(feature_values, target_values):
    """Calculate how well a feature separates classes"""
    means_by_class = []
    for class_val in np.unique(target_values):
        means_by_class.append(np.mean(feature_values[target_values == class_val]))
    return abs(means_by_class[1] - means_by_class[0])

age_separation = calculate_class_separation(X[:, 0], y)
salary_separation = calculate_class_separation(X[:, 1], y)

# Normalize separations
max_separation = max(age_separation, salary_separation)
age_importance = age_separation / max_separation
salary_importance = salary_separation / max_separation

features = ['Age', 'Salary']
importance_values = [age_importance, salary_importance]
colors = ['skyblue', 'lightcoral']

bars = plt.bar(features, importance_values, color=colors, alpha=0.7)
plt.ylabel('Relative Importance')
plt.title('Feature Importance Preview\n(Class Separation)')
plt.grid(True, alpha=0.3)

# Add values on bars
for bar, importance in zip(bars, importance_values):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.02,
             f'{importance:.3f}', ha='center', va='bottom')

# 6. Decision boundary analysis
plt.subplot(3, 4, 8)

# Create a simple grid to show potential splits
age_range = np.linspace(X[:, 0].min(), X[:, 0].max(), 50)
salary_range = np.linspace(X[:, 1].min(), X[:, 1].max(), 50)
Age_grid, Salary_grid = np.meshgrid(age_range, salary_range)

# Manual decision rule (preview of what tree might learn)
decision_surface = np.zeros_like(Age_grid)
for i in range(len(age_range)):
    for j in range(len(salary_range)):
        age_val = Age_grid[i, j]
        salary_val = Salary_grid[i, j]
        
        # Simple tree-like rules
        if age_val >= 35:
            if salary_val >= 60000:
                decision_surface[i, j] = 1  # Likely to purchase
            else:
                decision_surface[i, j] = 0  # Less likely
        else:
            if salary_val >= 50000:
                decision_surface[i, j] = 1
            else:
                decision_surface[i, j] = 0

plt.contourf(Age_grid, Salary_grid, decision_surface, alpha=0.3, levels=1, colors=['red', 'green'])
plt.scatter(X[y==0, 0], X[y==0, 1], c='red', alpha=0.6, s=30, label='No Purchase')
plt.scatter(X[y==1, 0], X[y==1, 1], c='green', alpha=0.6, s=30, label='Purchased')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Manual Decision Rules\n(Tree Preview)')
plt.legend()

# 7. Age groups analysis
plt.subplot(3, 4, 9)
age_bins = [18, 30, 40, 50, 60]
age_labels = ['18-29', '30-39', '40-49', '50-60']
age_groups = pd.cut(X[:, 0], bins=age_bins, labels=age_labels, include_lowest=True)

age_purchase_rates = []
for group in age_labels:
    mask = age_groups == group
    if mask.sum() > 0:
        purchase_rate = y[mask].mean()
        age_purchase_rates.append(purchase_rate)
    else:
        age_purchase_rates.append(0)

plt.bar(age_labels, age_purchase_rates, color='lightblue', alpha=0.7)
plt.ylabel('Purchase Rate')
plt.xlabel('Age Groups')
plt.title('Purchase Rate by Age Group\n(Natural Splits)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Add values on bars
for i, rate in enumerate(age_purchase_rates):
    plt.text(i, rate + 0.02, f'{rate:.2f}', ha='center', va='bottom')

# 8. Salary groups analysis
plt.subplot(3, 4, 10)
salary_bins = [15000, 40000, 60000, 80000, 150000]
salary_labels = ['15k-40k', '40k-60k', '60k-80k', '80k-150k']
salary_groups = pd.cut(X[:, 1], bins=salary_bins, labels=salary_labels, include_lowest=True)

salary_purchase_rates = []
for group in salary_labels:
    mask = salary_groups == group
    if mask.sum() > 0:
        purchase_rate = y[mask].mean()
        salary_purchase_rates.append(purchase_rate)
    else:
        salary_purchase_rates.append(0)

plt.bar(salary_labels, salary_purchase_rates, color='lightcoral', alpha=0.7)
plt.ylabel('Purchase Rate')
plt.xlabel('Salary Groups')
plt.title('Purchase Rate by Salary Group\n(Natural Splits)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Add values on bars
for i, rate in enumerate(salary_purchase_rates):
    plt.text(i, rate + 0.02, f'{rate:.2f}', ha='center', va='bottom')

# 9. Combined analysis
plt.subplot(3, 4, 11)
# Create a 2x2 grid based on age and salary splits
age_threshold = 35
salary_threshold = 60000

quadrant_labels = ['Young & Low Salary', 'Young & High Salary', 
                  'Older & Low Salary', 'Older & High Salary']
quadrant_rates = []

# Calculate purchase rates for each quadrant
conditions = [
    (X[:, 0] < age_threshold) & (X[:, 1] < salary_threshold),   # Young & Low
    (X[:, 0] < age_threshold) & (X[:, 1] >= salary_threshold),  # Young & High
    (X[:, 0] >= age_threshold) & (X[:, 1] < salary_threshold),  # Older & Low
    (X[:, 0] >= age_threshold) & (X[:, 1] >= salary_threshold)  # Older & High
]

for condition in conditions:
    if condition.sum() > 0:
        rate = y[condition].mean()
        quadrant_rates.append(rate)
    else:
        quadrant_rates.append(0)

colors = ['red', 'orange', 'yellow', 'green']
bars = plt.bar(range(len(quadrant_labels)), quadrant_rates, color=colors, alpha=0.7)
plt.ylabel('Purchase Rate')
plt.xlabel('Customer Segments')
plt.title('Purchase Rates by Segments\n(Tree Node Analysis)')
plt.xticks(range(len(quadrant_labels)), quadrant_labels, rotation=45, ha='right')
plt.grid(True, alpha=0.3)

# Add values on bars
for bar, rate in zip(bars, quadrant_rates):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.02,
             f'{rate:.2f}', ha='center', va='bottom')

# 10. Data quality summary
plt.subplot(3, 4, 12)
plt.text(0.1, 0.9, "Decision Tree Data Quality", fontsize=14, fontweight='bold', 
         transform=plt.gca().transAxes)

quality_text = f"""
TREE READINESS ASSESSMENT:

‚úÖ STRENGTHS:
‚Ä¢ No missing values
‚Ä¢ Continuous features (good for splits)
‚Ä¢ Clear class separation visible
‚Ä¢ Reasonable class balance ({class_counts[0]}/{class_counts[1]})
‚Ä¢ Multiple potential split points

üìä FEATURE ANALYSIS:
‚Ä¢ Age range: {X[:, 0].min():.0f}-{X[:, 0].max():.0f} years
‚Ä¢ Salary range: ${X[:, 1].min():.0f}-${X[:, 1].max():.0f}
‚Ä¢ Age separation: {age_separation:.1f} years
‚Ä¢ Salary separation: ${salary_separation:.0f}

üå≥ TREE EXPECTATIONS:
‚Ä¢ Primary splits likely on Age ~35 and Salary ~60k
‚Ä¢ Rectangular decision boundaries
‚Ä¢ High interpretability expected
‚Ä¢ Good performance likely
"""

plt.text(0.05, 0.05, quality_text, fontsize=9, fontfamily='monospace',
         transform=plt.gca().transAxes, verticalalignment='bottom',
         bbox=dict(boxstyle="round,pad=0.5", facecolor='lightgreen', alpha=0.7))

plt.axis('off')

plt.tight_layout()
plt.show()

# Summary insights for decision trees
print(f"\nüå≥ DECISION TREE INSIGHTS:")
print(f"   üìä Optimal split candidates:")
print(f"      ‚Ä¢ Age: Around 35 years (separates different life stages)")
print(f"      ‚Ä¢ Salary: Around $60,000 (separates economic segments)")
print(f"   üéØ Expected tree structure:")
print(f"      ‚Ä¢ Root: Likely age or salary split")
print(f"      ‚Ä¢ Depth: 2-3 levels should be sufficient")
print(f"      ‚Ä¢ Leaves: 4-8 terminal nodes expected")
print(f"   üìà Data quality: Excellent for tree learning")

print(f"\nüéØ Ready for Information Theory deep dive!")
````

## üßÆ Information Theory: The Mathematical Engine of Decision Trees

**Understanding Entropy, Information Gain, and Gini Impurity - The Core Concepts Behind Tree Construction**

Before we build Decision Trees, we need to understand the mathematical principles that guide how trees make optimal splitting decisions. This is where **Information Theory** comes into play!

### üéØ The Central Question
*"How do we choose the best feature and threshold to split our data at each node?"*

The answer lies in measuring **information content** and **uncertainty reduction**.

### üìä Key Concepts We'll Master

**1. üé≤ Entropy** - Measuring Uncertainty
- How "mixed" or "impure" is our data?
- Pure node (all same class) = Low entropy
- Mixed node (50/50 split) = High entropy

**2. üìà Information Gain** - Measuring Improvement
- How much does a split reduce uncertainty?
- Higher information gain = Better split
- Foundation of ID3 and C4.5 algorithms

**3. üéØ Gini Impurity** - Alternative Purity Measure
- Probability of misclassification
- Computationally faster than entropy
- Used in CART algorithm (scikit-learn default)

**4. üîç Splitting Criteria** - Choosing the Best Split
- Compare all possible splits
- Select the one with highest information gain
- Continue recursively until stopping criteria

````python
# Information Theory Implementation & Visualization
print("üßÆ INFORMATION THEORY FOR DECISION TREES")
print("="*50)

# Implement core information theory functions
def calculate_entropy(y):
    """
    Calculate Shannon entropy for a target variable
    Entropy = -Œ£(p_i * log2(p_i)) where p_i is probability of class i
    """
    if len(y) == 0:
        return 0
    
    # Count occurrences of each class
    _, counts = np.unique(y, return_counts=True)
    probabilities = counts / len(y)
    
    # Calculate entropy
    entropy = 0
    for p in probabilities:
        if p > 0:  # Avoid log(0)
            entropy -= p * log2(p)
    
    return entropy

def calculate_gini_impurity(y):
    """
    Calculate Gini impurity for a target variable
    Gini = 1 - Œ£(p_i^2) where p_i is probability of class i
    """
    if len(y) == 0:
        return 0
    
    # Count occurrences of each class
    _, counts = np.unique(y, return_counts=True)
    probabilities = counts / len(y)
    
    # Calculate Gini impurity
    gini = 1 - sum(p**2 for p in probabilities)
    
    return gini

def calculate_information_gain(parent_y, left_y, right_y, criterion='entropy'):
    """
    Calculate information gain from a split
    Information Gain = Parent_Impurity - Weighted_Average(Child_Impurities)
    """
    if criterion == 'entropy':
        parent_impurity = calculate_entropy(parent_y)
        left_impurity = calculate_entropy(left_y)
        right_impurity = calculate_entropy(right_y)
    else:  # gini
        parent_impurity = calculate_gini_impurity(parent_y)
        left_impurity = calculate_gini_impurity(left_y)
        right_impurity = calculate_gini_impurity(right_y)
    
    # Calculate weighted average of child impurities
    n_parent = len(parent_y)
    n_left = len(left_y)
    n_right = len(right_y)
    
    if n_parent == 0:
        return 0
    
    weighted_child_impurity = (n_left/n_parent) * left_impurity + (n_right/n_parent) * right_impurity
    
    # Information gain is the reduction in impurity
    information_gain = parent_impurity - weighted_child_impurity
    
    return information_gain

# Demonstrate with our dataset
print("üéØ INFORMATION THEORY DEMONSTRATION")
print("-"*40)

# Calculate initial entropy/gini of our dataset
initial_entropy = calculate_entropy(y)
initial_gini = calculate_gini_impurity(y)

print(f"üìä INITIAL DATASET IMPURITY:")
print(f"   üé≤ Entropy: {initial_entropy:.4f} bits")
print(f"   üéØ Gini Impurity: {initial_gini:.4f}")

# Analyze different potential splits
print(f"\nüîç ANALYZING POTENTIAL SPLITS:")

# Test different age thresholds
age_thresholds = [25, 30, 35, 40, 45, 50]
age_gains_entropy = []
age_gains_gini = []

print(f"\nüìà AGE THRESHOLD ANALYSIS:")
for threshold in age_thresholds:
    # Split data based on threshold
    left_mask = X[:, 0] <= threshold
    right_mask = X[:, 0] > threshold
    
    left_y = y[left_mask]
    right_y = y[right_mask]
    
    # Calculate information gains
    gain_entropy = calculate_information_gain(y, left_y, right_y, 'entropy')
    gain_gini = calculate_information_gain(y, left_y, right_y, 'gini')
    
    age_gains_entropy.append(gain_entropy)
    age_gains_gini.append(gain_gini)
    
    print(f"   Age ‚â§ {threshold}: Entropy Gain = {gain_entropy:.4f}, Gini Gain = {gain_gini:.4f}")

# Test different salary thresholds
salary_thresholds = [30000, 40000, 50000, 60000, 70000, 80000]
salary_gains_entropy = []
salary_gains_gini = []

print(f"\nüí∞ SALARY THRESHOLD ANALYSIS:")
for threshold in salary_thresholds:
    # Split data based on threshold
    left_mask = X[:, 1] <= threshold
    right_mask = X[:, 1] > threshold
    
    left_y = y[left_mask]
    right_y = y[right_mask]
    
    # Calculate information gains
    gain_entropy = calculate_information_gain(y, left_y, right_y, 'entropy')
    gain_gini = calculate_information_gain(y, left_y, right_y, 'gini')
    
    salary_gains_entropy.append(gain_entropy)
    salary_gains_gini.append(gain_gini)
    
    print(f"   Salary ‚â§ ${threshold:,}: Entropy Gain = {gain_entropy:.4f}, Gini Gain = {gain_gini:.4f}")

# Find optimal splits
best_age_idx_entropy = np.argmax(age_gains_entropy)
best_age_idx_gini = np.argmax(age_gains_gini)
best_salary_idx_entropy = np.argmax(salary_gains_entropy)
best_salary_idx_gini = np.argmax(salary_gains_gini)

print(f"\nüèÜ OPTIMAL SPLITS IDENTIFIED:")
print(f"   üìà Entropy Criterion:")
print(f"      ‚Ä¢ Best Age Split: ‚â§ {age_thresholds[best_age_idx_entropy]} (Gain: {age_gains_entropy[best_age_idx_entropy]:.4f})")
print(f"      ‚Ä¢ Best Salary Split: ‚â§ ${salary_thresholds[best_salary_idx_entropy]:,} (Gain: {salary_gains_entropy[best_salary_idx_entropy]:.4f})")
print(f"   üéØ Gini Criterion:")
print(f"      ‚Ä¢ Best Age Split: ‚â§ {age_thresholds[best_age_idx_gini]} (Gain: {age_gains_gini[best_age_idx_gini]:.4f})")
print(f"      ‚Ä¢ Best Salary Split: ‚â§ ${salary_thresholds[best_salary_idx_gini]:,} (Gain: {salary_gains_gini[best_salary_idx_gini]:.4f})")

# Determine overall best split
all_gains_entropy = age_gains_entropy + salary_gains_entropy
all_gains_gini = age_gains_gini + salary_gains_gini
all_thresholds = [(f"Age ‚â§ {t}", t) for t in age_thresholds] + [(f"Salary ‚â§ ${t:,}", t) for t in salary_thresholds]

best_overall_entropy_idx = np.argmax(all_gains_entropy)
best_overall_gini_idx = np.argmax(all_gains_gini)

print(f"\nüéØ ROOT NODE SPLIT RECOMMENDATION:")
print(f"   üìà Entropy: {all_thresholds[best_overall_entropy_idx][0]} (Gain: {all_gains_entropy[best_overall_entropy_idx]:.4f})")
print(f"   üéØ Gini: {all_thresholds[best_overall_gini_idx][0]} (Gain: {all_gains_gini[best_overall_gini_idx]:.4f})")

# Visualize information theory concepts
plt.figure(figsize=(20, 15))

# 1. Entropy vs Gini comparison for different class distributions
plt.subplot(3, 4, 1)
p_values = np.linspace(0.01, 0.99, 100)  # Probability of class 1
entropy_values = [-p*log2(p) - (1-p)*log2(1-p) for p in p_values]
gini_values = [2*p*(1-p) for p in p_values]

plt.plot(p_values, entropy_values, 'b-', linewidth=2, label='Entropy')
plt.plot(p_values, gini_values, 'r-', linewidth=2, label='Gini Impurity')
plt.xlabel('Probability of Class 1')
plt.ylabel('Impurity')
plt.title('Entropy vs Gini Impurity\n(Theoretical Comparison)')
plt.legend()
plt.grid(True, alpha=0.3)

# Mark key points
plt.axvline(0.5, color='gray', linestyle='--', alpha=0.7, label='Maximum Impurity')
plt.text(0.5, 0.8, 'Max Impurity\n(50/50 split)', ha='center', va='bottom')

# 2. Information gain by age threshold
plt.subplot(3, 4, 2)
plt.plot(age_thresholds, age_gains_entropy, 'bo-', linewidth=2, markersize=8, label='Entropy')
plt.plot(age_thresholds, age_gains_gini, 'ro-', linewidth=2, markersize=8, label='Gini')
plt.xlabel('Age Threshold')
plt.ylabel('Information Gain')
plt.title('Information Gain vs Age Threshold')
plt.legend()
plt.grid(True, alpha=0.3)

# Mark best splits
best_age_entropy = age_thresholds[best_age_idx_entropy]
best_age_gini = age_thresholds[best_age_idx_gini]
plt.axvline(best_age_entropy, color='blue', linestyle='--', alpha=0.7)
plt.axvline(best_age_gini, color='red', linestyle='--', alpha=0.7)

# 3. Information gain by salary threshold
plt.subplot(3, 4, 3)
plt.plot(salary_thresholds, salary_gains_entropy, 'bo-', linewidth=2, markersize=8, label='Entropy')
plt.plot(salary_thresholds, salary_gains_gini, 'ro-', linewidth=2, markersize=8, label='Gini')
plt.xlabel('Salary Threshold')
plt.ylabel('Information Gain')
plt.title('Information Gain vs Salary Threshold')
plt.legend()
plt.grid(True, alpha=0.3)

# Mark best splits
best_salary_entropy = salary_thresholds[best_salary_idx_entropy]
best_salary_gini = salary_thresholds[best_salary_idx_gini]
plt.axvline(best_salary_entropy, color='blue', linestyle='--', alpha=0.7)
plt.axvline(best_salary_gini, color='red', linestyle='--', alpha=0.7)

# 4. Split quality comparison
plt.subplot(3, 4, 4)
split_labels = [f"Age‚â§{t}" for t in age_thresholds] + [f"Sal‚â§{t//1000}k" for t in salary_thresholds]
all_gains = all_gains_entropy

bars = plt.bar(range(len(all_gains)), all_gains, 
               color=['lightblue']*len(age_thresholds) + ['lightcoral']*len(salary_thresholds),
               alpha=0.7)
plt.xlabel('Split Options')
plt.ylabel('Information Gain (Entropy)')
plt.title('All Split Options Comparison')
plt.xticks(range(len(split_labels)), split_labels, rotation=45, ha='right')
plt.grid(True, alpha=0.3)

# Highlight best split
best_bar = bars[best_overall_entropy_idx]
best_bar.set_color('gold')
best_bar.set_alpha(1.0)

# 5. Entropy calculation demonstration
plt.subplot(3, 4, 5)
# Show entropy for different class distributions
class_distributions = [
    ([100, 0], "Pure (100/0)"),
    ([90, 10], "Mostly Pure (90/10)"),
    ([70, 30], "Impure (70/30)"),
    ([50, 50], "Maximum Impurity (50/50)"),
    ([30, 70], "Impure (30/70)"),
    ([10, 90], "Mostly Pure (10/90)")
]

entropies = []
labels = []
for (counts, label) in class_distributions:
    total = sum(counts)
    probs = [c/total for c in counts]
    entropy = -sum(p*log2(p) if p > 0 else 0 for p in probs)
    entropies.append(entropy)
    labels.append(label)

colors = plt.cm.RdYlBu_r(np.linspace(0.2, 0.8, len(entropies)))
bars = plt.bar(range(len(entropies)), entropies, color=colors, alpha=0.8)
plt.xlabel('Class Distribution')
plt.ylabel('Entropy (bits)')
plt.title('Entropy for Different\nClass Distributions')
plt.xticks(range(len(labels)), [l.split('(')[1].replace(')', '') for l in labels], rotation=45)
plt.grid(True, alpha=0.3)

# Add values on bars
for bar, entropy in zip(bars, entropies):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.02,
             f'{entropy:.2f}', ha='center', va='bottom')

# 6. Gini impurity demonstration
plt.subplot(3, 4, 6)
ginis = []
for (counts, label) in class_distributions:
    total = sum(counts)
    probs = [c/total for c in counts]
    gini = 1 - sum(p**2 for p in probs)
    ginis.append(gini)

bars = plt.bar(range(len(ginis)), ginis, color=colors, alpha=0.8)
plt.xlabel('Class Distribution')
plt.ylabel('Gini Impurity')
plt.title('Gini Impurity for Different\nClass Distributions')
plt.xticks(range(len(labels)), [l.split('(')[1].replace(')', '') for l in labels], rotation=45)
plt.grid(True, alpha=0.3)

# Add values on bars
for bar, gini in zip(bars, ginis):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{gini:.2f}', ha='center', va='bottom')

# 7. Decision tree splitting visualization
plt.subplot(3, 4, 7)
# Show how the best split divides the data
best_threshold = age_thresholds[best_age_idx_entropy] if max(age_gains_entropy) > max(salary_gains_entropy) else salary_thresholds[best_salary_idx_entropy]
is_age_split = max(age_gains_entropy) > max(salary_gains_entropy)

if is_age_split:
    left_mask = X[:, 0] <= best_threshold
    right_mask = X[:, 0] > best_threshold
    split_text = f"Age ‚â§ {best_threshold}"
else:
    left_mask = X[:, 1] <= best_threshold
    right_mask = X[:, 1] > best_threshold
    split_text = f"Salary ‚â§ ${best_threshold:,}"

# Plot the split
plt.scatter(X[left_mask & (y==0), 0], X[left_mask & (y==0), 1], 
           c='red', alpha=0.6, s=30, marker='o', label='Left: No Purchase')
plt.scatter(X[left_mask & (y==1), 0], X[left_mask & (y==1), 1], 
           c='darkred', alpha=0.6, s=30, marker='^', label='Left: Purchased')
plt.scatter(X[right_mask & (y==0), 0], X[right_mask & (y==0), 1], 
           c='lightblue', alpha=0.6, s=30, marker='o', label='Right: No Purchase')
plt.scatter(X[right_mask & (y==1), 0], X[right_mask & (y==1), 1], 
           c='blue', alpha=0.6, s=30, marker='^', label='Right: Purchased')

if is_age_split:
    plt.axvline(best_threshold, color='black', linewidth=3, label=f'Split: {split_text}')
else:
    plt.axhline(best_threshold, color='black', linewidth=3, label=f'Split: {split_text}')

plt.xlabel('Age')
plt.ylabel('Salary')
plt.title(f'Optimal Split Visualization\n{split_text}')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

# 8. Information gain formula visualization
plt.subplot(3, 4, 8)
plt.text(0.1, 0.9, "Information Gain Formula", fontsize=12, fontweight='bold', 
         transform=plt.gca().transAxes)

formula_text = f"""
ENTROPY:
H(S) = -Œ£ p(c) √ó log‚ÇÇ(p(c))

INFORMATION GAIN:
IG(S,A) = H(S) - Œ£ (|S·µ•|/|S|) √ó H(S·µ•)

Where:
‚Ä¢ S = dataset
‚Ä¢ A = attribute (feature)
‚Ä¢ S·µ• = subset for value v
‚Ä¢ p(c) = proportion of class c

EXAMPLE CALCULATION:
Initial Entropy: {initial_entropy:.3f}
Best Split: {split_text}
Information Gain: {max(all_gains_entropy):.3f}

This means we reduce uncertainty
by {max(all_gains_entropy):.3f} bits!
"""

plt.text(0.05, 0.05, formula_text, fontsize=9, fontfamily='monospace',
         transform=plt.gca().transAxes, verticalalignment='bottom',
         bbox=dict(boxstyle="round,pad=0.5", facecolor='lightyellow', alpha=0.8))

plt.axis('off')

# 9-12: Detailed split analysis for top candidates
for i, (threshold, gain_entropy, gain_gini) in enumerate([(age_thresholds[best_age_idx_entropy], age_gains_entropy[best_age_idx_entropy], age_gains_gini[best_age_idx_gini]),
                                                          (salary_thresholds[best_salary_idx_entropy], salary_gains_entropy[best_salary_idx_entropy], salary_gains_gini[best_salary_idx_gini])]):
    plt.subplot(3, 4, 9 + i)
    
    if i == 0:  # Age split
        left_mask = X[:, 0] <= threshold
        title = f"Age Split ‚â§ {threshold}"
        feature_name = "Age"
    else:  # Salary split
        left_mask = X[:, 1] <= threshold
        title = f"Salary Split ‚â§ ${threshold:,}"
        feature_name = "Salary"
    
    right_mask = ~left_mask
    
    # Calculate split statistics
    left_classes = np.bincount(y[left_mask], minlength=2)
    right_classes = np.bincount(y[right_mask], minlength=2)
    
    # Create side-by-side bar chart
    x_pos = np.arange(2)
    width = 0.35
    
    plt.bar(x_pos - width/2, left_classes, width, label='Left Branch', alpha=0.7, color='lightcoral')
    plt.bar(x_pos + width/2, right_classes, width, label='Right Branch', alpha=0.7, color='lightblue')
    
    plt.xlabel('Class (0=No, 1=Yes)')
    plt.ylabel('Count')
    plt.title(f'{title}\nEntropy Gain: {gain_entropy:.3f}')
    plt.xticks(x_pos, ['No Purchase', 'Purchased'])
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    # Add purity information
    left_purity = max(left_classes) / sum(left_classes) if sum(left_classes) > 0 else 0
    right_purity = max(right_classes) / sum(right_classes) if sum(right_classes) > 0 else 0
    
    plt.text(0.5, 0.95, f'Left Purity: {left_purity:.2f}\nRight Purity: {right_purity:.2f}', 
             transform=plt.gca().transAxes, ha='center', va='top',
             bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

# Add remaining subplots for completeness
for i in range(11, 13):
    plt.subplot(3, 4, i)
    plt.text(0.5, 0.5, f"Additional Analysis\nSpace {i-10}", ha='center', va='center',
             transform=plt.gca().transAxes, fontsize=12)
    plt.axis('off')

plt.tight_layout()
plt.show()

print(f"\nüß† INFORMATION THEORY KEY INSIGHTS:")
print(f"   üìä Entropy measures uncertainty (0 = pure, 1 = maximum uncertainty)")
print(f"   üéØ Gini impurity measures misclassification probability")
print(f"   üìà Information gain measures uncertainty reduction from splits")
print(f"   üèÜ Best split: {all_thresholds[best_overall_entropy_idx][0]} with gain {all_gains_entropy[best_overall_entropy_idx]:.4f}")
print(f"   üå≥ This will be the root node of our decision tree!")

print(f"\nüéØ Ready for tree construction theory!")
````
